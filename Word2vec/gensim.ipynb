```python
file = "/Users/max/Desktop/Code/Microsoft_sentence_completion/Word2vec/relations.json"

with open(file, 'r') as instream:
    rel = json.load(instream)

import operator

class country_Capital():

    def __init__(self, wordvec):
        """wordvec is a group of vector word embeddings"""
        
        self.word_vec = wordvec

        self.ent = rel["capital-common-countries"]
        random.shuffle(self.ent)
        self.training_city, self.training_country = self.ent[0]

        #self.cities, self.countries = zip(*self.ent[1:])

    def predict_country(self):
        """Predicts the capital of countries in the list using the relation in the list"""
        output = []
        print("Training on {},{}".format(self.training_city, self.training_country))
        
        for city,country in self.ent[1:]:
            poss_country = self.word_vec.most_similar(positive = [self.training_country, city], negative = self.training_city)
            pred_country = sorted(poss_country, key = operator.itemgetter(1), reverse= True)[0][0]   
            tup = city, pred_country, country
            output.append(tup)
            print (tup)

        return output
    
    def predict_capital(self):
        """Predicts the capital of countries in the list using the relation in the list"""
        output = []
        print("Training on {},{}".format(self.training_city, self.training_country))
        for city,country in self.ent[1:]:
            
            poss_city = self.word_vec.most_similar(positive = [self.training_country, city], negative = self.training_city)
            pred_city = sorted(poss_city, key = operator.itemgetter(1), reverse= True)[0][0]            
            tup = (city, pred_city, country)
            output.append(tup)
            
            print (tup)
        return output

    def accuracy(self, method = "capital"):

        scores = []
        
        #city,capital = zip(*self.ent[1:])

        if method == "capital":
            pred_capital = self.predict_capital()
            
            for actual_capital,capital,_ in pred_capital:
                if capital == actual_capital:
                        scores.append(1)
                else:
                        scores.append(0)
                
        elif method == "country": 
            pred_country = self.predict_country()

            for _,country,goldcountry  in pred_country:
                if country == goldcountry:
                    scores.append(1)
                else:
                    scores.append(0)
                
        else:
            print("Not implemented".format(method))
         
        return sum(scores)/len(scores)

    def evaluate(self, runs = 10, method = "capital"):
        accuracy = []
        while runs > 0:
            accuracy.append(self.accuracy(method = method))
            runs -= 1

        mean_accuracy = np.mean(accuracy)
        std_accuracy = np.std(accuracy)

        return mean_accuracy,std_accuracy

    #def assess(self):

        #accuracy_capital = self.accuracy(method = "capital")              
        #accuracy_country = self.accuracy(method = "country")

        #return accuracy_capital, accuracy_country
```
